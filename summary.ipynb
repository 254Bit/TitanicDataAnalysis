```python
# A technical summary of the DataFrame
df.info()
```

```python
# The average age of the titanic passengers
df['Age'].mean()
```

```python
# The median age and ticket fare price of the Titanic Passengers
df[['Age','Fare']].median()
```

```python
# Describe
df[['Age','Fare']].describe()
```

```python
# Aggregation statistics can be calculated on entire columns or rows
df.agg({
    'Age': ['min','max','median','skew'],
    'Fare': ['min','max','median','skew'],
})
```

```python
# Average age for male versus female Titanic Passengers
# groupby provides the power of the split-apply-combine pattern
df[['Sex', 'Age']].groupby('Sex').mean()
```

```python
# To get the average age for each gender
#df.groupby('Sex').mean(numeric_only=True)
df.groupby('Sex')['Age'].mean()
```

```python
# Calculates the mean ticket fare price for each of the gender and cabin class combinations
df.groupby(['Sex','Pclass'])['Fare'].mean()
```

```python
# Number of passengers in each of the cabin classes
# The value_counts() method counts the number of records for each 
df['Pclass'].value_counts()
```

```python
#  Sort the Data according to the age of the passengers
df.sort_values(by='Age').head()
```

```python
# Sort the Titanic Data according to the cabin class and age in descending order
#dataframe.sort_values(), the rows in the table are sorted according to the defined column(s)
df.sort_values(by=['Pclass', 'Age'], ascending=False).head()
```

```python
# Make all name characters lowercase
df['Name'].str.lower()
```

```python
# Create a new column Surname that contains the surname of the passengers by extracting the part before the comma
df['Name'].str.split(',')
```

```python
# Using the Series.str.split() method, each of the values is returned as a list of 2 elements. The first element is the part before the comma and the second element is the part after the comma.
df['Surname'] = df['Name'].str.split(',').str.get(0)
df['Surname']
```

```python
# Extract the passenger data about the Mr on board the Titanic
df['Name'].str.contains('Mr')
```

```python
# In the Sex column, replace values of "male" by "M" and values of Female by "F"
df['gender_shortened'] = df['Sex'].replace({'male':'M', 'female':'F'})
df['gender_shortened']
```
